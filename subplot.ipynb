```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x)
```

```python
print(y)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'k')
plt.subplot(2,1,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.plot(y,x)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```
