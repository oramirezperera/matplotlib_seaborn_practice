```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'r')
ax2.plot(y, x, 'b')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), 'r')
axes[0,2].plot(x, np.tan(x), 'g')
axes[0,3].plot(x, np.cos(x)**2, 'k')
```

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'r')
ax3.plot(x,np.tan(x),'g')
ax4.plot(x,np.cos(x)**2,'k')
ax5.plot(y,np.cos(y))
ax6.plot(y,np.sin(y),'r')
ax7.plot(y,np.tan(y),'g')
ax8.plot(y,np.cos(y)**2,'k')
fig.tight_layout()
```
