```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'bo')
plt.show()
```

```python
plt.plot(x,y, 'r--')
plt.show()
```

```python
plt.plot(x,y, 'rx')
plt.show()
```

```python
plt.plot(x,y, 'ys-')
plt.show()
```

```python
plt.plot(x,y, 'gD:')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.plot(x,y,'g-.')
plt.show()
```
