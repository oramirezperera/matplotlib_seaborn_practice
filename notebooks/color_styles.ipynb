```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+1, 'g--')
ax.plot(x, x+2, 'r-')
ax.plot(x, x+3, 'b.-')
ax.plot(x, x+4, 'y:')
plt.show()
```

Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+1, color = 'green', alpha=0.5, linewidth=5, linestyle='-', marker='o') # with apha I can maje transparency
ax.plot(x, x+2, color = 'blue', alpha=0.4, linewidth=8, linestyle='dashed', marker='x', markersize=10, markeredgecolor='#FF0000')
ax.plot(x, x+3, color='#66FF66', alpha=0.8, linewidth=3, linestyle='-.',marker='v')
ax.plot(x, x+4, color='#CCFFCC', linewidth=2, linestyle=':',marker='P')
plt.show()
```
