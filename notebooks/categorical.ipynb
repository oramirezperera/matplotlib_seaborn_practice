```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, palette='dark', marker='o')
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box', dodge=True, col='time')
plt.show()
```
