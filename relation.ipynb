```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.15, 0.5))
plt.show()
```

```python
markers = {"Lunch": "s", "Dinner": "D"}
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='day', style='time', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.15, 0.5))
plt.show()
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip", hue='day')
plt.legend(loc='center', bbox_to_anchor=(1.15, 0.5))
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='day', kind='line', col='time')
plt.show()
```
